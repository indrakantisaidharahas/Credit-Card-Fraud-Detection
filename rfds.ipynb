```python
from sklearn.utils import resample

df_majority = df[df.Class == 0]
df_minority = df[df.Class == 1]

df_majority_downsampled = resample(df_majority,
                                  replace=False,
                                  n_samples=1000,
                                  random_state=42)

df_balanced = pd.concat([df_majority_downsampled, df_minority])
df_balanced["Class"].value_counts()
```

```python
import matplotlib.pyplot as plt
from itertools import combinations

cols = df.columns

# take only PCA columns (skip Time, Amount, Class if needed)
pcs = [col for col in cols if col.startswith('V')]

clas = [0, 1]
colors = ['red', 'blue']

# generate all pairs
for col1, col2 in combinations(pcs, 2):
    plt.figure(figsize=(5,4))
    
    for i, k in zip(clas, colors):
        subset = df[df['Class'] == i]
        plt.scatter(subset[col1], subset[col2], color=k, label=f"Class {i}", alpha=0.3)

    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.title(f"{col1} vs {col2}")
    plt.legend()
    plt.show()
```
